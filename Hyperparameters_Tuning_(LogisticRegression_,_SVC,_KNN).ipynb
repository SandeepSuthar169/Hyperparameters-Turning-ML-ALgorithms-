# GridSearcgCV


```python
# 1. LogisticRegression
```

```python
penalty = ['l1', 'l2','elasticnet']
dual =[False]
tol = [1e-4]
C = [0.01, 0.1, 1.0]
class_weight = [None, 'balanced']
solver=['lbfgs', 'liblinear', 'sag', 'saga']
random_state = [None]
max_iter = [10, 50, 100]
```

```python
param_grid = {
    'penalty':penalty,
    'dual':dual,
    'C':C,
    'class_weight':class_weight,
    'solver':solver,
    'random_state':random_state,
    'max_iter':max_iter
}

print(param_grid)
```

```python
from sklearn.model_selection import GridSearchCV

lr = LogisticRegression()

lr_grid = GridSearchCV(estimator = lr,
                             param_grid=param_grid,
                             cv = 5,
                             verbose=2,
                             n_jobs = -1)
```

```python
lr_grid.fit(x_train, y_train)
```

```python
lr_grid.best_params_
```

```python
lr_grid.best_score_
```

```python
# 2. SVM (Support Vector Macgine)
```

```python
C = [1.0]
kernel = ['linear', 'poly', 'rdf', 'sigmoid']
degree = [1, 2, 3]
gamma = ['scale', 'auto']
coef0 = [0.0001, 0.0]
class_weight =['balance', None]
#decison_function_shape = ['ovo', 'ovr']
```

```python
param_grid = {
    'C':C,
    'kernel':kernel,
    'degree':degree,
    'gamma':gamma,
    'coef0':coef0,
    'class_weight':class_weight,
    #'decison_function_shape':decison_function_shape
}

print(param_grid)
```

```python
from sklearn.model_selection import GridSearchCV
svc = SVC()

svc_grid = GridSearchCV(estimator = svc,
                             param_grid =param_grid,
                             cv = 5,
                             verbose=2,
                             n_jobs = -1)
```

```python
svc_grid
```

```python
svc_grid.fit(x_train, y_train)
```

```python
svc_grid.best_params_
```

```python
svc_grid.best_score_
```

```python
# 3. KNN Classifier
```

```python
n_neighbors = [3, 5, 7, 10]
metrics = ['minkowski']
p = [1, 2]
weights = ['uniform', 'distance']
```

```python
param_grid ={
    'n_neighbors':n_neighbors,
    'metric':metrics,
    'p':p,
    'weights':weights
}

print(param_grid)
```
